```

import pandas as pd 
import matplotlib.pyplot as plt
```

```
data= pd.read_csv('http://13.234.66.67/summer19/datasets/bank.csv')
```

```

data.head(5)
```

```

data.tail(5)
```

```

plt.scatter(data.Age,data.CreditScore,label="Age",marker='x')
plt.xlabel('Age')
plt.ylabel('CreditScore')
```

```
plt.bar(data.Age,data.EstimatedSalary)
plt.xlabel('Age')
plt.ylabel('EstimatedSalary')
```

```
plt.bar(data.CustomerId,data.CreditScore)
plt.xlabel('CustomerId')
plt.ylabel('CreditScore')
```

```
plt.bar(data.Age,data.Balance)
plt.xlabel('Age')
plt.ylabel('Balance')
```

```
plt.stackplot(data.Age,data.Balance)
plt.xlabel('Age')
plt.ylabel('Balance')
```
