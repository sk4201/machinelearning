```
import matplotlib.pyplot as plt

import numpy as np
import time
```

```
x=[4,7,8,9,11,5]
y=[13,23,5,7,9,3]
player=['virat','dhoni','vijay','dhawan']
runs=[120,44,77,33]
exp=[0.1,0,0,0.2]
```

```
#p;otting graph
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y)    #ploting line only
plt.scatter(x,y)
plt.bar(x,y, label="pollution")
plt.bar(player,runs,label="cricketscore")
plt.grid(color='green')
plt.show()
```

```

plt.scatter(x,y)
plt.pie(runs ,labels=player, explode=exp, shadow=True, autopct='%1.1f%%')
plt.grid(color='green')

plt.show()
```
